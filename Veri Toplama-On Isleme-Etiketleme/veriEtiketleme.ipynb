```
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline
import torch
import pandas as pd
import numpy as np
model = AutoModelForSequenceClassification.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
tokenizer = AutoTokenizer.from_pretrained("savasy/bert-base-turkish-sentiment-cased")
sa = pipeline("sentiment-analysis", tokenizer=tokenizer, model=model)
input_file='/content/drive/MyDrive/veri/181307025_veriKumesi.csv') 
```

```
data['sentiments'] = ''
for i in range(len(data)):
    comment = data['comments'][i]
    p=sa(comment)
    deger=p[0]['label']
    if (deger=='positive'):
      etiket=1;
    
    else:
      etiket=0;

    data['sentiments'][i]=etiket
    print("Index: ", i ,  "sentiments" ,etiket)
```

```
data.to_csv('/content/drive/MyDrive/veri/etiketliDataset.csv',encoding='utf-8-sig',index=False)
```
